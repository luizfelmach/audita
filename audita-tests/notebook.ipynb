```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./result.csv")
df
```

```python
dfs = [df_sub for _, df_sub in df.groupby("extra_tags")]
len(dfs)
```

```python
df = dfs[0]
df_filtrado = df[df["metric_name"] == "http_reqs"]
df_filtrado
```

```python


# Somar os valores para cada timestamp
df_grouped = df_filtrado.groupby("timestamp")["metric_value"].sum()

# Criar o grÃ¡fico
plt.plot(df_grouped.index, df_grouped.values, marker="o", linestyle="-")

# ConfiguraÃ§Ãµes do grÃ¡fico
plt.xlabel("Timestamp")
plt.ylabel("Soma dos Valores")
plt.title(f"Soma da MÃ©trica 'http_reqs' ao Longo do Tempo")
plt.xticks(rotation=45)
plt.grid()

plt.show()
```

```python
plt.figure(figsize=(10, 5))  # Define o tamanho do grÃ¡fico

# Percorre todos os DataFrames na lista dfs
for i, df in enumerate(dfs):
    # Filtrar apenas a mÃ©trica desejada
    df_filtrado = df[df["metric_name"] == "http_reqs"]

    # Somar os valores por timestamp
    df_grouped = df_filtrado.groupby("timestamp")["metric_value"].sum()

    label = df_filtrado["extra_tags"].iloc[0] if not df_filtrado["extra_tags"].empty else f"Dataset {i+1}"

    # Adicionar ao grÃ¡fico
    plt.plot(df_grouped.index, df_grouped.values, marker="o", linestyle="-", label=f"Dataset {label}")

# ConfiguraÃ§Ãµes do grÃ¡fico
plt.xlabel("Timestamp")
plt.ylabel("Soma dos Valores")
plt.title("Soma da MÃ©trica 'http_reqs' ao Longo do Tempo")
plt.xticks(rotation=45)
plt.legend(title="Conjunto de Dados")  # Adiciona legenda para identificar cada linha
plt.grid()

# Exibir o grÃ¡fico
plt.show()
```

```python
plt.figure(figsize=(10, 5))  # Define o tamanho do grÃ¡fico

# Percorre todos os DataFrames na lista dfs
for i, df in enumerate(dfs):
    # Filtrar apenas a mÃ©trica desejada
    df_filtrado = df[df["metric_name"] == "http_reqs"]

    # Aplicar o shift para alinhar os timestamps
    df_filtrado["timestamp"] = df_filtrado["timestamp"] - df_filtrado["timestamp"].min()

    # Somar os valores por timestamp
    df_grouped = df_filtrado.groupby("timestamp")["metric_value"].sum()

    label = df_filtrado["extra_tags"].iloc[0] if not df_filtrado["extra_tags"].empty else f"Dataset {i+1}"

    # Adicionar ao grÃ¡fico
    plt.plot(df_grouped.index, df_grouped.values, marker="o", linestyle="-", label=f"Dataset {label}")

# ConfiguraÃ§Ãµes do grÃ¡fico
plt.xlabel("Timestamp (Alinhado)")
plt.ylabel("Soma dos Valores")
plt.title("Soma da MÃ©trica 'http_reqs' ao Longo do Tempo (Timestamps Alinhados)")
plt.xticks(rotation=45)
plt.legend(title="Conjunto de Dados")  # Adiciona legenda para identificar cada linha
plt.grid()

# Exibir o grÃ¡fico
plt.show()
```

```python

plt.figure(figsize=(10, 5))  # Define o tamanho do grÃ¡fico

# Percorre todos os DataFrames na lista dfs
for i, df in enumerate(dfs):
    # Filtrar apenas a mÃ©trica de tempo de espera
    df_filtrado = df[df["metric_name"] == "http_req_waiting"]

    # Aplicar o shift para alinhar os timestamps
    df_filtrado["timestamp"] = df_filtrado["timestamp"] - df_filtrado["timestamp"].min()

    # Calcular a mÃ©dia do tempo de espera por timestamp (ou outra agregaÃ§Ã£o, como soma)
    df_grouped = df_filtrado.groupby("timestamp")["metric_value"].mean()  # Ou .sum(), dependendo da sua necessidade

    label = df_filtrado["extra_tags"].iloc[0] if not df_filtrado["extra_tags"].empty else f"Dataset {i+1}"

    # Adicionar ao grÃ¡fico
    plt.plot(df_grouped.index, df_grouped.values, marker="o", linestyle="-", label=f"Dataset {label}")

# ConfiguraÃ§Ãµes do grÃ¡fico
plt.xlabel("Timestamp (Alinhado)")
plt.ylabel("Tempo de Espera (ms)")
plt.title("Tempo de Espera ('http_req_waiting') ao Longo do Tempo (Timestamps Alinhados)")
plt.xticks(rotation=45)
plt.legend(title="Conjunto de Dados")  # Adiciona legenda para identificar cada linha
plt.grid()

# Exibir o grÃ¡fico
plt.show()
```
